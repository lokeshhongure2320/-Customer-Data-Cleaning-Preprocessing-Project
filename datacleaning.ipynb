```python
df2=df
```

```python
for i in df2:
    plt.boxplot(df2[i])
    plt.title(i)
    plt.show()
```

```python
for i in df2:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    upper_tail=q3+1.5*iqr
    lower_tail=q1-1.5*iqr

    print(i)
    print("upper_tail: ",upper_tail)
    print("lower_tail: ",lower_tail)
    print("---------------------------------------------------------------")
    
```

```python
df.loc[df['Fresh']>upper_tail,'Fresh']=upper_tail
df.loc[df['Milk']>upper_tail,'Milk']=upper_tail
df.loc[df['Grocery']>upper_tail,'Grocery']=upper_tail
df.loc[df['Frozen']>upper_tail,'Frozen']=upper_tail
df.loc[df['Detergents_Paper']>upper_tail,'Detergents_Paper']=upper_tail
df.loc[df['Delicatessen']>upper_tail,'Delicatessen']=upper_tail
```

```python
df.loc[df['Fresh']<lower_tail,'Fresh']=lower_tail
```

```python
for i in df:
    plt.boxplot(df[i])
    plt.title(i)
    plt.show()
```
